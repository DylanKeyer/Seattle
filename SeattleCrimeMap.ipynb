```python
import json
import pandas as pd
import urllib
from pandas.io.json import json_normalize
import folium
from pandas_ply import *
install_ply(pd)
import geocoder
```

```python
url = 'https://data.seattle.gov/resource/7ais-f98f.json?$limit=7000&$where=year>2013'
text = (urllib.urlopen(url)).read()
```

```python
data = json_normalize(json.loads(text))
crimes = data.ply_select('date_reported','latitude','longitude','year','summarized_offense_description')
```

```python
g = geocoder.google('Seattle, WA')
g.latlng
```

```python
sMap = folium.Map(location=g.latlng,zoom_start=12)
```

```python
for crime in crimes.iterrows():
    sMap.simple_marker(location=[crime[1]['latitude'],crime[1]['longitude']], popup=crime[1]['summarized_offense_description'], clustered_marker=True)
```
